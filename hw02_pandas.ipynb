**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
%%time
counter_by_group = dict(zip(list(set(df["race/ethnicity"])),[0,0,0,0,0]))
reading_sum_by_group = dict(zip(list(set(df["race/ethnicity"])),[0,0,0,0,0]))
writing_list_by_group = dict(zip(list(set(df["education"])),[[],[],[],[],[],[]]))
for i in range(len(df)):
  row = df[["race/ethnicity", "education", "reading score", "writing score"]].iloc[i,:]
  counter_by_group[row[0]] += 1
  reading_sum_by_group[row[0]] += row[2]
  writing_list_by_group[row[1]].append(row[3])
reading_mean_by_group = {}
writing_min_by_group = {}
for key in reading_sum_by_group.keys():
  reading_mean_by_group[key] = reading_sum_by_group[key] / counter_by_group[key]
for key in writing_list_by_group.keys():
    writing_min_by_group[key] = min(writing_list_by_group[key])
print(pd.DataFrame([reading_mean_by_group]).T.sort_index(),"\n",  pd.DataFrame([writing_min_by_group]).T.sort_index())
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
def pass_or_fail(score):
  return "F" if score < passmark else "P"
df["Math_PassStatus"] = df["math score"].map(pass_or_fail)
df["Reading_PassStatus"] = df["reading score"].map(pass_or_fail)
df["Writing_PassStatus"] = df["writing score"].map(pass_or_fail)
print("Сдало экзамен по математике: ", df["Math_PassStatus"].value_counts()[0],
      "\nНе сдало экзамен по математике:", df["Math_PassStatus"].value_counts()[1],
      "\nСдало экзамен по reading:",df["Reading_PassStatus"].value_counts()[0],
      "\nНе сдало экзамен по reading:",df["Reading_PassStatus"].value_counts()[1],
      "\nСдало экзамен по writing:",df["Writing_PassStatus"].value_counts()[0],
      "\nНе сдало экзамен по writing:", df["Writing_PassStatus"].value_counts()[1])
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Посчитайте количество студентов, которые сдали все экзамены.

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark >= 90:
        return 'A'
    elif average_mark >= 80:
        return 'B'
    elif average_mark >= 70:
        return 'C'
    elif average_mark >= 60:
        return 'D'
    elif average_mark >= 50:
        return 'E'
    else:
        return 'F'
df["Grade"] = ((df["math score"] + df["reading score"] + df["writing score"])/3).apply(GetGrade)
df["Grade"].value_counts(sort=True, ascending = True)
```
